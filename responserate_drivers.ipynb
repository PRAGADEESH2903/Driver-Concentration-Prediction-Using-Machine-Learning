```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, cohen_kappa_score, classification_report

# Assuming X_train, X_test, y_train, y_test are already defined

# Initialize a dictionary to store performance metrics
performance_metrics = {}

# Iterate through each model in the dictionary
for name, model in model_dict.items():
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Calculate performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')  # Weighted for multiclass
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    kappa = cohen_kappa_score(y_test, y_pred)
    
    # Store the metrics in the performance_metrics dictionary
    performance_metrics[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Kappa': kappa
    }
    
    # Display metrics for each model
    print(f"======================================{name}=======================================")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Cohen's Kappa: {kappa:.4f}")
    print("\n")
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from itertools import cycle

# Binarize the output (One-vs-Rest strategy)
y_test_binarized = label_binarize(y_test, classes=sorted(set(y_test)))  # Ensure classes are sorted
n_classes = y_test_binarized.shape[1]

# Initialize dictionaries to hold False Positive Rates (fpr), True Positive Rates (tpr), and AUC scores for each model
fpr_dict = {}
tpr_dict = {}
roc_auc_dict = {}

# Iterate through each model in the dictionary
for name, model in model_dict.items():
    # Check if the model has the predict_proba method
    if hasattr(model, "predict_proba"):
        # Use predict_proba if available
        y_pred_proba = model.predict_proba(X_test)
    else:
        # Use predict and manually convert labels to one-hot encoding
        y_pred = model.predict(X_test)
        y_pred_proba = label_binarize(y_pred, classes=sorted(set(y_test)))

    # Calculate the False Positive Rate, True Positive Rate, and AUC score for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_pred_proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Store the values in the dictionaries for each model
    fpr_dict[name] = fpr
    tpr_dict[name] = tpr
    roc_auc_dict[name] = roc_auc

    # Plot ROC curves for each class for the current model
    plt.figure(figsize=(10, 8))
    colors = cycle(["aqua", "darkorange", "cornflowerblue"])
    for i, color in zip(range(n_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label=f'Class {i} (AUC = {roc_auc[i]:.2f})')
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curves for {name}")
    plt.legend(loc="lower right")
    plt.show()
```
